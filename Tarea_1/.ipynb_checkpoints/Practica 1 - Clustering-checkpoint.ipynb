```python
def get_kmeans_clusters_centroides(k, data, iteraciones):    
    rows = data.shape[0]    
    cols = data.shape[1]
    clusters = {}
    centroides = np.array([]).reshape(cols, 0)
    
    for i in range(k):
        random_centroids = random.randint(0, rows-1)
        centroides = np.c_[centroides, data.iloc[random_centroids]]
        
    for i in range(iteraciones):
        dist = np.array([]).reshape(rows, 0)
        for i in range(k):
            temp_dist = np.sum((data - centroides[:, i])**2, axis = 1)
            dist = np.c_[dist, temp_dist]
        min_dist_index = np.argmin(dist, axis=1) + 1
        
        clusters = {}
        
        for i in range(k):
            clusters[i + 1] = np.array([]).reshape(cols, 0)
        
        for i in range(rows):
            clusters[min_dist_index[i]] = np.c_[clusters[min_dist_index[i]], data.iloc[i]]
            
        for i in range(k):
            clusters[i + 1] = clusters[i + 1].T
            
        for i in range(k):
            centroides[:, i] = np.mean(clusters[i + 1], axis=0)
            
    return (clusters, centroides)
            
import matplotlib.cm as cm
def plot_kmeans_clusters(clusters, centroides):
    colors = ['red', 'green', 'blue', 'yellow', 'black']
    names = ['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4', 'Cluster 5']
    
    
    
    
    
    
    
    
    for i in range(k):
        plt.scatter(clusters[i + 1][:, 0], clusters[i + 1][:, 1], c = colors[i], label = names[i])
        
    plt.scatter(centroides[0, :], centroides[1, :], s = 300, c = 'purple', label = 'Centroides')
    plt.title('Clusterizacion')
    plt.xlabel('Estatura')
    plt.ylabel('Edad')
    plt.legend()
    plt.show()
```

##### Experimentos

```python
k = 1
clusters, centroides = get_kmeans_clusters_centroides(k, data_normales, 100)
plot_kmeans_clusters(clusters, centroides)

k = 2
clusters, centroides = get_kmeans_clusters_centroides(k, data_normales, 100)
plot_kmeans_clusters(clusters, centroides)

k = 3
clusters, centroides = get_kmeans_clusters_centroides(k, data_normales, 100)
plot_kmeans_clusters(clusters, centroides)

k = 4
clusters, centroides = get_kmeans_clusters_centroides(k, data_normales, 100)
plot_kmeans_clusters(clusters, centroides)

k = 5
clusters, centroides = get_kmeans_clusters_centroides(k, data_normales, 100)
plot_kmeans_clusters(clusters, centroides)
```

### Soft-clustering con GMM

```python
def plot_gmm_clusters(data, gmm):
    clusters = gmm.predict(data)
    
    names = ['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4', 'Cluster 5']
    
    plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c = clusters)
    
    plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], s = 300, c = 'black', label = 'Centroides')
    plt.title('Clusterizacion GMM')
    plt.xlabel('Estatura')
    plt.ylabel('Edad')
    plt.legend()
    plt.show()
    
k1 = 1
k2 = 2
k3 = 3
k4 = 4
k5 = 5
gmm1 = GaussianMixture(n_components = k1, random_state = 0).fit(data_normales)
gmm2 = GaussianMixture(n_components = k2, random_state = 0).fit(data_normales)
gmm3 = GaussianMixture(n_components = k3, random_state = 0).fit(data_normales)
gmm4 = GaussianMixture(n_components = k4, random_state = 0).fit(data_normales)
gmm5 = GaussianMixture(n_components = k5, random_state = 0).fit(data_normales)

plot_gmm_clusters(data_normales, gmm1)
plot_gmm_clusters(data_normales, gmm2)
plot_gmm_clusters(data_normales, gmm3)
plot_gmm_clusters(data_normales, gmm4)
plot_gmm_clusters(data_normales, gmm5)
```

### Analisis metodo del codo

```python
# M辿todo codo para identificar el n炭mero 坦ptimo de clusters
# ==============================================================================
from sklearn.cluster import KMeans
range_n_clusters = range(1, 15)
inertias = []

for n_clusters in range_n_clusters:
    modelo_kmeans = KMeans(
                        n_clusters   = n_clusters, 
                        n_init       = 20, 
                        random_state = 123
                    )
    modelo_kmeans.fit(data_normales)
    inertias.append(modelo_kmeans.inertia_)

fig, ax = plt.subplots(1, 1, figsize=(6, 3.84))
ax.plot(range_n_clusters, inertias, marker='o')
ax.set_title("Evoluci坦n de la varianza intra-cluster total")
ax.set_xlabel('N炭mero clusters')
ax.set_ylabel('Intra-cluster (inertia)');
```

#### Conclusion
A partir de 3 clusters la reduccion en la suma total de cuadrados internos parece estabilizarse, indicando que k = 3 es un buen numero de clusters.

```python
def mc_pi_aprox(N=1000):
    plt.figure(figsize=(8,8))  # tama単o de la figura
    x, y = np.random.uniform(-1, 1, size=(2, N))
    interior = (x**2 + y**2) <= 1
    pi = interior.sum() * 4 / N
    error = abs((pi - np.pi) / pi) * 100
    exterior = np.invert(interior)
    plt.plot(x[interior], y[interior], 'b.')
    plt.plot(x[exterior], y[exterior], 'r.')
    plt.plot(0, 0, label='$\hat \pi$ = {:4.4f}\nerror = {:4.4f}%'
             .format(pi,error), alpha=0)
    plt.axis('square')
    plt.legend(frameon=True, framealpha=0.9, fontsize=16)

mc_pi_aprox()
```
